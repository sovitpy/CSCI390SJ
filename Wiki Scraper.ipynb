```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_lynching_victims_in_the_United_States')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
final_data = []
tables = soup.find_all('table', {'class':'wikitable'})
myTable = tables[1] # change the index of this list to grab different tables (tables[0] is the first table tables[1] is the second and so on...)
headers = myTable.find_all('th')
headers = list(map(lambda x : x.text.strip(), headers))
rows = myTable.find_all('tr')
for i in range(1, len(rows)):
    row = rows[i]
    data = row.find_all('td')
    data = list(map(lambda x : x.text.strip(), data))
    data_tuple = zip(headers, data)
    data_dict = {}
    for data_point in data_tuple:
        data_dict[data_point[0]] = data_point[1]
    final_data.append(data_dict)
```

```python
final_data
```

```python
import pandas as pd
df = pd.DataFrame(final_data)
```

```python
df.to_csv('wiki_1900-1909.csv')
```
