```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
charge_count = df.groupby(['Charge'])['Charge'].count()
```

```python
charge_count_df = charge_count.to_frame()
```

```python
x_values = list(charge_count_df.index)
y_values = list(charge_count_df['Charge'])
```

```python
ax = sns.barplot(x = y_values, y = x_values, orient='h')
ax.set_yticklabels(ax.get_yticklabels(), fontsize=8)
ax.set_xlabel('\nNumber of Cases',fontsize=12)
ax.set_ylabel('\nCharges',fontsize=12)
ax.set_title("\nNumber of Cases by Charges in 1893\n", fontsize=15)
plt.show()
```
